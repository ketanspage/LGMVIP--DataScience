```python
from google.colab.patches import cv2_imshow
import cv2
```

## ***reading the image***

```python
image=cv2.imread('Deadpool.jpg')#reading the image
cv2_imshow(image)
```

## ***grayscale of the loaded image***

```python
grayscale=cv2.cvtColor(image,cv2.COLOR_RGB2GRAY)#converting the image into grayscale(black and white)
cv2_imshow(grayscale)
```

```python
blur=cv2.GaussianBlur(grayscale,(21,21),0)#bluring the grayscale 
cv2_imshow(blur)
```

## ***pencil sketch***

```python
pencil_sketch=cv2.divide(grayscale,blur,scale=256.0)#dividing the blured grayscale with grayscale
cv2_imshow(pencil_sketch)
```

## ***filling colors in the pencil sketch***

```python
colored_blur=cv2.GaussianBlur(image,(21,21),0)
colored_sketch=cv2.divide(image,colored_blur,scale=160.0)
cv2_imshow(colored_sketch)
```

## ***cartooon portrait of the image*** 

```python
gray=cv2.medianBlur(grayscale,5)
edges=cv2.adaptiveThreshold(gray,255,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY,9,9)
color=cv2.bilateralFilter(image,9,250,250)
cartoon=cv2.bitwise_and(color,color,mask=edges)
cv2_imshow(cartoon)
```
