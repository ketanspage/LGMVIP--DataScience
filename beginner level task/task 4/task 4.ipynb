```python
import cv2
import sys
```

```python
image=cv2.imread("Deadpool.jpg")#reading the image
```

```python
if image is None:
    sys.exit("could not read the image")
else:
    cv2.imshow("fam",image)
    cv2.waitKey(0)
```

```python
grayscale=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)#converting the image to grayscale(negetive image)
cv2.imshow("fam(-ve)",grayscale)
cv2.waitKey(0)
```

```python
invert_grayscale=255-grayscale#inverting the grayscale image
cv2.imshow("fam_inverted",invert_grayscale)
cv2.waitKey(0)
```

```python
blurred = cv2.GaussianBlur(invert_grayscale, (21, 21), 0)#blurring the inverted image 
inverted_blurred = 255 - blurred#inverting the blurred image
cv2.imshow("jdo",inverted_blurred)
cv2.waitKey(0)
```

```python
pencil_sketch = cv2.divide(grayscale, inverted_blurred, scale=256.0)#dividing the grayscale
cv2.imshow("Sketch", pencil_sketch)
cv2.waitKey(0)
```

```python
import matplotlib.pyplot as plt
plt.imshow(pencil_sketch)
plt.show()
```
